```python
class MovieHandler( xml.sax.ContentHandler ):
    def __init__(self):
        self.CurrentData = ""
        self.further_information = ""
        self.locations = ""
        self.year = ""
        self.rating = ""
        self.stars = ""
        self.description = ""
 
    # 元素开始事件处理
    def startElement(self, tag, attributes):
        self.CurrentData = tag
        if tag == "record":
            print("*****record*****")
            category = attributes["category"]
            #print("category:", category)

    # 元素结束事件处理
    def endElement(self, tag):
        if self.CurrentData == "location":
            print("location:", self.locations)
        self.CurrentData = ""
 
    # 内容事件处理
    def characters(self, content):
        if self.CurrentData == "location":
            self.locations = content
  
```
